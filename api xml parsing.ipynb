## df2
- 유기날짜(검색시작) : 20190101
- 유기날짜(검색 종료일) : 20191231

```python
# 동물보호관리시스템 유기동물 조회 서비스 API

url = 'http://openapi.animal.go.kr/openapi/service/rest/abandonmentPublicSrvc/abandonmentPublic'

payload = {
    'serviceKey': '1zQWNFsLekxRSL9eOgAGYY+0gq339y3pEyBW1vC31LkJf48SZvduvflcNjBC3/Oej9jwqY40e+7yDKdK+gDRRw==',
    'bgnde': '20190101',
    'endde': '20191231',
    'upr_cd': '6110000',
    'numOfRows': 7500
}

response = requests.get(url, params=payload)

print(response.text)
```

```python
import xml.etree.ElementTree as ET

# XML 데이터 파싱 

root = ET.fromstring(response.text) 
```

```python
# XML 데이터를 dictionary의 list로 만들기

rows = []

for item in root.iter('item'):
    row = {}
    for child in list(item):
        row[child.tag] = child.text #child에 tag랑 text 사용할 수 있데
    rows.append(row)

print('갯수:', len(rows))

print(rows[0])
```

```python
# 첫번째 item을 얻습니다.

item = next(root.iter('item'))

item
```

```python
# item 바로 아래에 뭐가 있을까?

list(item)
```

```python
# 컬럼 목록 준비
columns = []

# Accumulation
for child in list(item):
    columns.append(child.tag)

child

# 컬럼 목록 확인
print(columns)
```

```python
# DataFrame으로 만들기

df2 = pd.DataFrame(rows, columns=columns)


df2
```

```python
df2.to_csv('./df2.csv', index=True, encoding='CP949')
```
